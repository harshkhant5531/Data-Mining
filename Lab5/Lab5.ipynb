### 3)  Apply Scaling to AGE attribute with min max, decimal scaling and z score.


```python
cpy4=df.copy()
cpy4
```

```python
#standard scaling -> Z-score 
```

```python
#formula of Max-Min of normalization
# x'=(x-xmin)/(xmax-xmin)

df['age_maxmin']=(df['Age']-df['Age'].min())/(df['Age'].max()-df['Age'].min())
df['age_maxmin']
```

```python
#standard scaling -> Z-score 

# z=(x-xmean)/std

df['age_maxminZ']=(df['Age']-df['Age'].mean())/(df['Age'].std())
df['age_maxminZ']
```

```python
#decimal scaling

#x'=x/(10^j)  
#j=max value of age column then find digit count

cpy5=df.copy()
cpy5

cpy5['Age']=cpy5['Age'].interpolate()

j=len(str(int(cpy5['Age'].max())))
print(j)

#x'=x/(10^j) 

df['age_maxminDS']=(df['Age'])/(10**j)
df['age_maxminDS']
```
